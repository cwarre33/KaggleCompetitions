# Dataset Pipeline Creator badge

This notebook **writes output files** so you can create a dataset from notebook output and earn the [Dataset Pipeline Creator](https://www.kaggle.com/progression/badges) badge.

## Steps
1. **Run all cells** (they will write files to `/kaggle/working/`).
2. In the **right-hand panel**, open the **Output** or **Data** tab for this notebook.
3. Find the option **"Create dataset from notebook output"** or **"New dataset from output"** and click it.
4. Add a title/description and create the dataset. Badge earned.

```python
# On Kaggle, writable directory is /kaggle/working/
import os
import pandas as pd

out_dir = "/kaggle/working"
if not os.path.exists(out_dir):
    out_dir = "."  # fallback when run locally

# Create a small output dataset
df = pd.DataFrame({
    "id": [1, 2, 3, 4, 5],
    "value": [10.5, 20.3, 30.1, 40.0, 50.2],
    "label": ["A", "B", "A", "B", "A"],
})
path = os.path.join(out_dir, "pipeline_output.csv")
df.to_csv(path, index=False)
print(f"Written: {path}")
```

```python
# Optional: write another file so the output clearly has "data" to package
readme = os.path.join(out_dir, "README.txt")
with open(readme, "w") as f:
    f.write("Dataset created from notebook output for the Dataset Pipeline Creator badge.")
print(f"Written: {readme}")
```

After running the cells above, use the **Output** panel on the right → **Create dataset from notebook output** (or similar) to create your dataset and earn the badge.
