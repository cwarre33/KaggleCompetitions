# Model Pipeline Creator badge

This notebook **saves a model to the notebook output** so you can create a model from it and earn the [Model Pipeline Creator](https://www.kaggle.com/progression/badges) badge.

## Steps
1. **Run all cells** — they will write model files to `/kaggle/working/`.
2. In the **right-hand panel**, open the **Output** or **Data** tab.
3. Find **"Create model from notebook output"** / **"New model from output"** and use it to create a model from this notebook's output.
4. Badge earned.

```python
import os
import pickle

out_dir = "/kaggle/working"
if not os.path.exists(out_dir):
    out_dir = "."

# Minimal "model" artifact (a small serialized object)
model_artifact = {"version": 1, "type": "pipeline-badge-demo", "params": {}}
path = os.path.join(out_dir, "model.pkl")
with open(path, "wb") as f:
    pickle.dump(model_artifact, f)
print(f"Saved: {path}")
```

```python
# Optional: add a second file so the output clearly has model-like files
readme = os.path.join(out_dir, "README.txt")
with open(readme, "w") as f:
    f.write("Model output for Model Pipeline Creator badge. Create a model from this notebook output.")
print(f"Saved: {readme}")
```

After running the cells, use the **Output** panel → **Create model from notebook output** (or similar) to create your model and earn the badge.
